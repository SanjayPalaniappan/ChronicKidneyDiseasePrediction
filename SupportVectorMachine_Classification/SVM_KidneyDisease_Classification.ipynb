```python
#Importing the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Reading the dataset
ckd_dataset=pd.read_csv('CKD.csv')
ckd_dataset
```

```python
#Details about the dataset
ckd_dataset.info()
```

```python
#Dimensions of dataset
ckd_dataset.shape
```

```python
#Splitting input and output columns
X = ckd_dataset.iloc[:,:-1]
y = ckd_dataset['classification']
```

```python
#Printing X
X
```

```python
#Printing y
y
```

```python
#Converting categorical column values into numerical values
ckd_dataset = pd.get_dummies(X, drop_first=True)
ckd_dataset
```

```python
#Getting the column names
ckd_dataset.columns
```

```python
#Getting number of rows and number of columns
ckd_dataset.shape
```

```python
#split into training set and test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(ckd_dataset, y, test_size = 1/3, random_state = 0)
```

```python
#Printing X_train
X_train
```

```python
#Standardizing data to bring all column values into same scale
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
#Performing grid search for support vector machine
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
param_grid = {'kernel':['linear','rbf','poly','sigmoid'],
             'gamma':['auto','scale'],
             'C':[10,500,1000,3000,5000]} 
grid = GridSearchCV(SVC(probability=True), param_grid, refit = True, verbose = 3,n_jobs=-1,scoring='f1_weighted') 
# fitting the model for grid search 
grid.fit(X_train, y_train) 
```

```python
# print best parameter after tuning 
print(grid.best_params_) 
re=grid.cv_results_
grid_predictions = grid.predict(X_test) 
   

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, grid_predictions)
cm


# print classification report 
from sklearn.metrics import classification_report
clf_report = classification_report(y_test, grid_predictions)
clf_report
```

```python
# Calculating f1 score
from sklearn.metrics import f1_score
f1_macro=f1_score(y_test,grid_predictions,average='weighted')
print("The f1_macro value for best parameter {}:".format(grid.best_params_),f1_macro)
```

```python
#Printing the confusion matrix
print("The confusion Matrix:\n",cm)
```

```python
#Printing the Classification report
print("The report:\n",clf_report)
```

```python
#Plotting ROC curve
from sklearn.metrics import roc_auc_score
roc_auc_score(y_test,grid.predict_proba(X_test)[:,1])
```
